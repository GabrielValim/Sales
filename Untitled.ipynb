```python
#importando bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
```

```python
#importando as basess

feature_df = pd.read_csv("Features.csv")
sales = pd.read_csv('sales_data-set.csv')
stores = pd.read_csv('stores_data-set.csv')
```

```python
#visualizando a feature
feature_df.info()
feature_df.head()
```

```python
#visualizando a sales
sales.info()
sales.head()
```

```python
#visualizando a stores
stores.info()
stores.head()
```

```python
#Transformando colunas em datas
feature_df["Date"] = pd.to_datetime(feature_df["Date"])
sales["Date"] = pd.to_datetime(sales["Date"])
```

```python
#juntando tabelas
df_merge = pd.merge(
    sales,
    feature_df,
    how='left',
    on =['Store','Date', 'IsHoliday']
)

df_merge = pd.merge(
    df_merge,
    stores,
    how='left',
    on ='Store'
)
df_merge = df_merge.fillna(0)

```

```python
#aplicando fun√ßoes de data
df_merge['ano']= df_merge['Date'].dt.year
df_merge['mes']= df_merge['Date'].dt.month
df_merge['dia']= df_merge['Date'].dt.day
```

## Visualizando de forma grafica

```python
top_stores = df_merge.groupby(by=['Type'], as_index=False)['Weekly_Sales'].sum()

fig,ax = plt.subplots(figsize=(10,5))

ax.set_title("Top Lojas",
             fontsize=20,
             fontweight="bold",
             color="#5d16de")

ax.tick_params(axis="x",labelsize=20)
ax.tick_params(axis="y",labelsize=16)

plt.bar(top_stores.Type,
        top_stores.Weekly_Sales,
        color="#5d16de")



plt.show()
```

```python
data_sales_month = df_merge.groupby("mes").agg({"Weekly_Sales":"sum"})
plt.figure(figsize=(10, 5))
sns.barplot(x=data_sales_month.index,y=data_sales_month.Weekly_Sales)
plt.title("vendas por mes")
plt.xlabel("mes")
plt.ylabel("vendas")
```

```python
data_sales_year = df_merge.groupby('ano').agg({"Weekly_Sales":"sum"})

sns.barplot(x=data_sales_year.index,
            y=data_sales_year.Weekly_Sales.sort_values(ascending=False))

plt.title("vendas por ano")
plt.xlabel("ano")
plt.ylabel("vendas")
```

```python
top_dpt = df_merge.groupby("Store")["Weekly_Sales"].sum()
top_dpt = top_dpt.reset_index()
top_dpt = top_dpt.sort_values("Weekly_Sales",ascending=False)

top_N_dpt = top_dpt.head(5)
```

### Feature Engineer

```python
#transformando coluna
df_merge["IsHoliday"].apply(lambda x: 1 if x == True else 0)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
#dividindo a coluna type
ohe = ohe.fit(df_merge[["Type"]])
ohe.transform(df_merge[["Type"]]).toarray()
```

```python
ohe_df = pd.DataFrame(ohe.transform(df_merge[["Type"]]).toarray())
ohe_df.head()
```

```python
ohe.get_feature_names_out()
```

```python
ohe_df.columns = ohe.get_feature_names_out()
ohe_df.head()
```

```python
df_merge = pd.concat([df_merge,ohe_df],axis=1)
df_merge.head()
```

```python
df_merge = df_merge.drop('Type', axis=1)
```

```python
df_merge = df_merge.drop("IsHoliday", axis=1)
```

### StandartScaler

```python
scaler = StandardScaler()
```

```python
scaler = scaler.fit(df_merge[['Weekly_Sales']])
```

```python
df_merge["Weekly_Sales_padrao"] = scaler.transform(df_merge[['Weekly_Sales']])
```

### RobustScaler

```python
scaler = RobustScaler()
```

```python
scaler = scaler.fit(df_merge[["Weekly_Sales"]])
```

```python
df_merge["Weekly_Sales_robust"] = scaler.transform(df_merge[["Weekly_Sales"]])
```

```python
#mostrando graficamente
fig,ax = plt.subplots(ncols=2,figsize=(15,6))

sns.kdeplot(df_merge.Weekly_Sales,ax=ax[0])
sns.kdeplot(df_merge.Weekly_Sales_padrao,ax=ax[1],label="padrao")
sns.kdeplot(df_merge.Weekly_Sales_robust,ax=ax[1],label="robust")

ax[1].legend()

plt.show();
```
